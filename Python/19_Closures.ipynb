# Closures
A closure is a function object that remembers values from its enclosing scope even after the outer function has finished executing.

A closure requires:
- A nested function
- The nested function uses a variable from the outer function
- The outer function returns the nested function

---

## Basic Example

```python
def outer(x):
    def inner():
        return x
    return inner

f = outer(10)
print(f())  # 10


````python
def outer_fun(a):
    print("Print the value of a ", a)
    def inner_fun(b):
        print("Print the vallue of b", b)
        return a+b
    return inner_fun
````

````python
closure = outer_fun(10)
````
