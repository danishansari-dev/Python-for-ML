# lambda function
    - anonymous function

```python
def add(a,b):
    return a+b
```

```python
add(4,5)
```

```python
lambda a,b : a+b
```

```python
c = lambda a,b : a+b
```

```python
c(3,4)
```

```python
square = lambda a : a**2
```

```python
square(6)
```

```python
max = lambda a,b : a if a>b else b
```

```python
max(56,34)
```

```python
even_odd = lambda number : "even" if number % 2 == 0 else "odd"
```

```python
even_odd(56)
```
