```python
def add(*dan): # takes as a tuple input
    print(dan)
    j = 0
    for i in dan:
        j += i
    return j
```

```python
add(3,4,5,66,3,2,3,345,2)
```

```python
def profile1(**dan):
    for i in dan.items():
        print(i)
```

```python
profile1(Name ="Danish", Batch = "Python")
```

```python
def profile(company, **dan):
    for i in dan.items():
        print(i,company)
```

```python
profile("euron", name = "sudh")
```
