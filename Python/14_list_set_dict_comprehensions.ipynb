# List Comprehensions

```python
l = []
for i in range(1,6):
    print(i**2)
    l.append(i**2)
```

```python
l
```

```python
l1 = [i**2 for i in range(1,6)] # comprehension
```

```python
l1
```

```python
[i for i in range(1,20) if i % 2 == 0]
```

```python
l = []
for i in range(20):
    if i%2 == 0 :
        l.append(i)
```

```python
l
```

```python
words = ["python", "Euron", "danish"]
[i.upper() for i in words]
```

```python
l = [[3,4], [6,7]]
l_final = [3,4,6,7]
```

```python
[i for i in l]
```

```python
[j for i in l for j in i]
```

```python
[i for i in range(20) if i%2 == 0 if i%3 == 0]
```

```python
{i for i in [4,5,6,76,46,7,7,8,6,4]}
```

```python
l = ["danish", "euron", "ansari", "python", "hindi"]
```

```python
{i[0] for i in l}
```

```python
{i:i**2 for i in range(5)}
```

```python
s = 'danish'
```

```python
{i:d for i,d in enumerate(s)}
```

```python
list(enumerate(s))
```

```python
l1 = [1,2,3]
l2 = ["danish", "ansari", "euron"]
```

```python
list(zip(l1,l2))
```

```python
{i:v for i,v in zip(l1,l2)}
```
