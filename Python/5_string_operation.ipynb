```python
from string import Template
```

```python
Template("my name is $name my age is $age").substitute(name=name, age=23
                                                       )
```

```python
email_template = Template("""
                          From: $sender
                          To $receiver
                          Subject: $subject
                          $body""")
```

```python
email_template.substitute(sender = "danishansari.dev@gmail.com", 
                          receiver="ashish@gmail.com", 
                          subject="Test Mail", 
                          body="This is a test mail")
```

```python
email_template.substitute(sender = input("danishansari.dev@gmail.com"), 
                          receiver="ashish@gmail.com", 
                          subject="Test Mail", 
                          body="This is a test mail")
```
