## Patterns

```python
n = 10
for i in range(n):
    print("*"*i)
```

```python
n = 10
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
n = 10
for i in range(1, n+1):
    print(" " *(n-i) + "*"*i)
```

```python
n =10
for i in range(1, n+1):
    print(" "*(n-i) + "*"*(2*i-1))
```

```python
n =6
for i in range(1, n+1):
    print(" "*(n-i) + "*"*(2*i-1))
for i in range(n-1,0,-1):
    print(" "*(n-i) + "*"*(2*i-1))
```
