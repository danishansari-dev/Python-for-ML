# NumPy Revision Notes

## Array Creation
- `np.array()` → create array
- `np.arange()` → range of values
- `np.zeros()` → zeros
- `np.ones()` → ones
- `np.eye()` → identity matrix
- `np.random.rand()` → random floats

---

## Array Properties
- `ndim` → number of dimensions
- `shape` → dimensions
- `size` → number of elements
- `dtype` → data type

---

## Reshaping
- `reshape()` → change shape
- `ravel()` → flatten (view)
- `flatten()` → flatten (copy)
- `expand_dims()` → add axis
- `transpose()` → transpose
- `swapaxes()` → swap axes

---

## Joining and Splitting
- `np.concatenate()` → join
- `np.vstack()` → vertical stack
- `np.hstack()` → horizontal stack
- `np.column_stack()` → stack columns
- `np.split()` → split
- `np.hsplit()` → horizontal split
- `np.vsplit()` → vertical split

---

## Sorting and Searching
- `np.sort()` → sort
- `np.argmax()` → index of max
- `np.where()` → conditional index
- `np.unique()` → unique values

---

## Math and Stats
- `np.sum()` → sum
- `np.mean()` → mean
- `np.max()` → max
- `np.min()` → min
- `np.std()` → standard deviation
- `np.var()` → variance
- `np.percentile()` → percentile
- `np.cumsum()` → cumulative sum
- `np.clip()` → limit values
- `np.corrcoef()` → correlation

---

## Modification
- `np.append()` → append
- `np.delete()` → delete
- `np.put()` → replace values
- `np.flip()` → reverse

---

## Linear Algebra
- `np.dot()` → dot product
- `np.matmul()` → matrix multiplication
- `np.linalg.inv()` → inverse
- `np.linalg.det()` → determinant

---

## Set Operations
- `np.union1d()` → union
- `np.intersect1d()` → intersection
- `np.setdiff1d()` → difference
- `np.setxor1d()` → symmetric difference
- `np.isin()` → membership

---

## Histogram
- `np.histogram()` → frequency distribution

---

## Broadcasting Rules
- Dimensions must match or be 1
- Smaller dimension is expanded
- If sizes mismatch and neither is 1 → error

---

## General Rules
- Operations are vectorized
- Original arrays are not modified unless assigned
- Views vs copies matter (`ravel` vs `flatten`)

---

## Summary
NumPy is built for fast numerical computation on multi-dimensional arrays using vectorized operations, broadcasting, and optimized C backend.
