## Array Operations

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
```

```python
# Scalar operations

# arithmetic
a1 ** 2
```

```python
# relational
a2 > 15
```

```python
# vector operations
# arithmetic */+-
a1 + a2
```

## Array Functions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
# max/min/sum/prod
# 0 -> col & 1 -> row
np.min(a1,axis=0)
```

```python
# mean/median/std/var
np.mean(a1)
```

```python
np.median(a1, axis=0)
```

```python
# dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

a3
```

```python
np.dot(a2,a3)
```

```python
# log and exponents
np.exp(a1)
```

```python
# round/floor/ceil
np.round(np.random.random((2,3))*100)
```
